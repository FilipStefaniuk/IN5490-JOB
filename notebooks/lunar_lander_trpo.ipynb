## TRPO Algorithm Specific Hyperparameters

### Max KL

```python
max_kl_experiments = {e.params['max_kl']: e for e in get_experiments(
                                  '../experiments/lunar_lander/trpo/max_kl/params/',
                                  '../experiments/lunar_lander/trpo/max_kl/results/1234',
                                  '../experiments/lunar_lander/trpo/max_kl/results/1313',
                                  '../experiments/lunar_lander/trpo/max_kl/results/1337')}

_, ax = plt.subplots(figsize=(14, 6))
plot_auc_2d(list(max_kl_experiments.values()), min_val=-200, max_val=200, xscale='log', ax=ax)
```
