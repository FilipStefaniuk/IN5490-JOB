## Common Hyperparameters with respect to max KL

### max KL - Gamma


```python
max_kl_gamma_experiments = {(e.params['max_kl'], e.params['gamma']) : e for e in get_experiments(
                                  '../experiments/bipedal_walker/trpo/max_kl_gamma/params/',
                                  '../experiments/bipedal_walker/trpo/max_kl_gamma/results/1234')}
param1 = sorted(set([i[0] for i in list(max_kl_gamma_experiments.keys())]))
param2 = sorted(set([i[1] for i in list(max_kl_gamma_experiments.keys())]))
interact(lambda max_kl, gamma : plot_learning_curve(max_kl_gamma_experiments[(max_kl, gamma)], aspect=2.3), max_kl=param1, gamma = param2);
```

```python
experiments = get_experiments('../experiments/bipedal_walker/trpo/max_kl_gamma/params/',
                              '../experiments/bipedal_walker/trpo/max_kl_gamma/results/1234')
plot_heatmap(experiments, min_val=-100, max_val=300)
plot_3d(experiments, True, min_val=-100, max_val=300)
```
