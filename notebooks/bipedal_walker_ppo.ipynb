```python
import os
import sys

sys.path.append(os.path.dirname(os.getcwd()))

from plotter import *
from ipywidgets import interact, fixed

import matplotlib.pyplot as plt
```

## PPO Algorithm Specific Hyperparameters

### Cliprange

```python
cliprange_experiments = {e.params['cliprange']: e for e in get_experiments(
                                  '../experiments/bipedal_walker/ppo/cliprange/params/',
                                  '../experiments/bipedal_walker/ppo/cliprange/results/1234',
                                  '../experiments/bipedal_walker/ppo/cliprange/results/1313',
                                  '../experiments/bipedal_walker/ppo/cliprange/results/1337')}
interact(lambda cliprange: plot_learning_curve(cliprange_experiments[cliprange], y='eprewmean', x='total_timesteps', aspect=2.3), cliprange=sorted(cliprange_experiments.keys()));
```

```python
experiments = get_experiments('../experiments/bipedal_walker/ppo/cliprange/params/',
                              '../experiments/bipedal_walker/ppo/cliprange/results/1234',
                              '../experiments/bipedal_walker/ppo/cliprange/results/1313',
                              '../experiments/bipedal_walker/ppo/cliprange/results/1337')
plot_auc_2d(experiments, min_val=-100, max_val=300, x='total_timesteps', y='eprewmean')
plt.savefig("bipedal_walkerppo_auc2d.png")
```

## Common Hyperparameters with respect to clipRange


### cliprange - Gamma

```python
cliprange_gamma_experiments = {(e.params['cliprange'], e.params['gamma']) : e for e in get_experiments(
                                  '../experiments/bipedal_walker/ppo/cliprange_gamma/params/',
                                  '../experiments/bipedal_walker/ppo/cliprange_gamma/results/1234')}
param1 = sorted(set([i[0] for i in list(cliprange_gamma_experiments.keys())]))
param2 = sorted(set([i[1] for i in list(cliprange_gamma_experiments.keys())]))
interact(lambda cliprange, gamma : plot_learning_curve(cliprange_gamma_experiments[(cliprange, gamma)], y='eprewmean', x='total_timesteps', aspect=2.3), cliprange=param1, gamma = param2);
```

```python
experiments = get_experiments('../experiments/bipedal_walker/ppo/cliprange_gamma/params/',
                              '../experiments/bipedal_walker/ppo/cliprange_gamma/results/1234')
plot_heatmap(experiments,x='total_timesteps', y='eprewmean', min_val=-100, max_val=300)
plt.savefig("bipedal_walkerppocliprange_gamma1234heatmap.png")
plot_3d(experiments, True, x='total_timesteps', y='eprewmean', min_val=-100, max_val=300)
plt.savefig("bipedal_walkerppocliprange_gamma1234plot3d.png")
```
